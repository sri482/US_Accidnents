```python
# Convert to datetime with error handling
data['Start_Time'] = pd.to_datetime(data['Start_Time'], errors='coerce', format='%Y-%m-%d %H:%M:%S', exact=False)

# Check for any conversion issues
invalid_dates = data[data['Start_Time'].isna()]

data.Start_Time[0]
```

```python
data.Start_Time.dt.hour
```

```python
sns.distplot(data.Start_Time.dt.hour,bins=24,kde=False,norm_hist=True)
```

  - A high percentage of accidents occur between 6AM to 10 AM (Probably people in a hurry to get work)
  -  Next zhighest percentage between 3PM to 6PM

sns.distplot(data.Start_Time.dt.dayofweek,bins=7,kde=False,norm_hist=True)

```python
sns.distplot(data.Start_Time.dt.dayofweek,bins=7,kde=False,norm_hist=True)
```

- Is the distribution of accidents by hour the same on weekends as week days

```python
sundays=data[data.Start_Time.dt.dayofweek==6]
sns.distplot(sundays.Start_Time.dt.hour,bins=24,kde=False,norm_hist=True)
```

```python
mondays=data[data.Start_Time.dt.dayofweek==0]
sns.distplot(mondays.Start_Time.dt.hour,bins=24,kde=False,norm_hist=True)
```

- On sundays, the peak occurs between 10AM to 3PM, unlike Weekdays

```python
data_2019=data[data.Start_Time.dt.year==2019]
data_2019_Bing=data_2019[data_2019.Source=='Bing']
sns.distplot(data_2019.Start_Time.dt.month,bins=12,kde=False,norm_hist=True)
```
